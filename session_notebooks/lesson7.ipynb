```python
def complement_base(base):
    """Returns the Watson-Crick complement of a base."""
    if base in 'Aa':
        return 'T'
    elif base in 'Tt':
        return 'A'
    elif base in 'Gg':
        return 'C'
    elif base in 'Cc':
        return 'G'
    else:
        return 'N'
```

```python
complement_base('N')
```

```python
complement_base('c')
```

```python
seq = 'GACAGACTCCATGNNNNNNNCACGTGGGUATCTGTC'
```

```python
complement_base(seq)
```

```python
def complement(seq):
    """Complement an entire sequence of DNA"""
    complement_seq = ''
    for base in seq:
        complement_seq += complement_base(base)
    return complement_seq
```

```python
complement(seq)
```

```python
def rev_complement(seq):
    """Reverse complement an entire sequence of DNA"""
    complement_seq = ''
    for base in reversed(seq):
        complement_seq += complement_base(base)
    return complement_seq
```

```python
def complement_base(base, moltype='DNA'):
    """Returns the Watson-Crick complement of a base.
    base: str - a single character base
    moltype: str - can be 'DNA' or 'RNA', the type of biomolecule
    """
    print('moltype is', moltype)
    if base in 'Aa':
        if moltype == 'DNA':
            return 'T'
        elif moltype == 'RNA':
            return 'U'
    elif base in 'TtUu':
        return 'A'
    elif base in 'Gg':
        return 'C'
    elif base in 'Cc':
        return 'G'
    else:
        return 'N'
```

```python
complement_base('U', moltype='RNA')
```

```python
complement_base('A')
```

```python
complement_base('A', moltype='DNA')
```
